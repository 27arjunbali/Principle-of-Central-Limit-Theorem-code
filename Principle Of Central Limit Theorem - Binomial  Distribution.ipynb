```python
import random
import math
import matplotlib.pyplot as plt
import numpy as np
from itertools import groupby
```

```python
def nck(N,K):
    f=math.factorial
    return f(N) // f(K) // f(N-K)    
```

```python
n=1000
N=30
CDFprob=0
p=0.7
q=(1-p)
l=[]
```

```python
for i in range(0,n):
    for j in range(0,N):
        x=0
        CDFprob=0
        u=random.uniform(0,1)
        while(CDFprob<=u):
            CDFprob+=nck(N,x)*(p**x)*(q**(N-x))
            x+=1
    l.append(x)
print(l)
```

```python
mean=[]
for i in l:
    mean.append(i/N)
```

```python
print(mean)
```

```python
mean.sort()
print(mean)
```

```python
unique=[]
count_val=1
frequency=[]
for i in range(0,(n-1)):
    if mean[i]!=mean[i+1]:
        count_val+=1
```

```python
for i in range(0,count_val):
    frequency.append(1)
```

```python
print(frequency)
```

```python
count_val=0
for i in range(0,(n-1)):
    if mean[i]==mean[i+1]:
        frequency[count_val]+=1
    else:
        count_val+=1
        unique.append(mean[i])
unique.append(mean[i])
```

```python
plt.bar(unique,frequency,0.025)
```
