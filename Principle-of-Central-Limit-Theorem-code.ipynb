```python
import random
import math
import matplotlib.pyplot as plt
import numpy as np
from itertools import groupby
```

```python
def nck(N,K):
    f=math.factorial
    return f(N) // f(K) // f(N-K)    
```

```python
n=1000
N=30
CDFprob=0
p=0.7
q=(1-p)
l=[]
```

```python
for i in range(0,n):
    list1=[]
    for j in range(0,N):
        x=0
        CDFprob=0
        u=random.uniform(0,1)
        while(CDFprob<=u):
            CDFprob+=(nck(N,x)*((p)**x)*((q)**(N-x)))
            x+=1
        list1.append(x)
    l.append(list1)
print(l)
```

```python
mean=[]
for i in l:
    mean.append(sum(i)/N)
```

```python
print(mean)
```

```python
mean.sort()
print(mean)
```

```python
unique=[]
count_val=1
frequency=[]
for i in range(0,(n-1)):
    if mean[i]!=mean[i+1]:
        count_val+=1
```

```python
for i in range(0,count_val):
    frequency.append(1)
```

```python
print(frequency)
```

```python
count_val=0
for i in range(0,(n-1)):
    if mean[i]==mean[i+1]:
        frequency[count_val]+=1
    else:
        count_val+=1
        unique.append(mean[i])
unique.append(mean[i])
```

```python
plt.bar(unique,frequency,0.05)
```

```python
lamda=0.7
n2=1000
N2=30
```

```python
def fact(f):
    fact=math.factorial(f)
    return fact
```

```python
p=[]
h=math.exp(-lamda)
for i in range(0,n2):
    p1=[]
    for j in range(0,N2):
        prob=0
        k=0
        v=random.uniform(0,1)
        while(prob<=v):
            prob+=(h*(lamda**k)/fact(k))
            k+=1
        p1.append(k)
    p.append(p1)
```

```python
print(p)
```

```python
meanp=[]
for i in p:
    meanp.append(sum(i)/N2)
```

```python
print(meanp)
```

```python
meanp.sort()
print(meanp)
```

```python
freq1=[]
cal=1
uniq=[]
for i in range(0,(n2-1)):
    if meanp[i]!=meanp[i+1]:
        cal+=1
```

```python
for i in range(0,cal):
    freq1.append(1)
```

```python
cal=0
for i in range(0,(n2-1)):
    if meanp[i]==meanp[i+1]:
        freq1[cal]+=1
    else:
        cal+=1
        uniq.append(meanp[i])
uniq.append(meanp[i+1])
```

```python
print(uniq)
```

```python
print(freq1)
```

```python
plt.bar(uniq,freq1,0.025)
```
